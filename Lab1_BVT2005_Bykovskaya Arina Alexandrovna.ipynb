# Задание №2
Написать генератор случайных матриц(многомерных), который принимает опциональные параметры m, n, min_limit, max_limit, где m и n указывают размер матрицы, а min_lim и max_lim - минимальное и максимальное значение для генерируемого числа.

```python
import random
import time
```

```python
from random import randint
from time import sleep

user_m = input()
user_n = input()
user_min_limit = input()
user_max_limit = input()

def print_mat(mat):
    for line in mat:
        out = [str(el) for el in line]
        print(" ".join(out))


def matrix_generator(m, n, min_lim, max_lim):
    return [[randint(int(min_lim), int(max_lim)) for j in range(int(n))] for i in range(int(m))]

seconds_left = 60
while seconds_left > 0:
    seconds_left -= 1
    generated_matrix = matrix_generator(user_m, user_n, user_min_limit, user_max_limit)
    print(seconds_left)
    print_mat(generated_matrix)
sleep(1)
```

```python
import copy
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
def selection_sort(line):
    for i in range(len(line)):
        temp = line[i]
        index = i
        for j in range(i, len(line)):
            if line[j] < temp:
                temp = line[j]
                index = j
        line[index], line[i] = line[i], temp


matrix = copy.deepcopy(generated_matrix)
t1 = time.time()
for line in matrix:
    line = selection_sort(line)
t2 = time.time()

selection_sort_time = round((t2 - t1) * 1000)

print_mat(matrix)
print(selection_sort_time)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# сортировка пузырьком

start_time = time.time()
def bubble_sort(line):
    n = 1
    for i in range (len(line) - 1):
        for j in range(len(line) - n):
            if line[j] > line[j + 1]:
                line[j], line[j + 1] = line[j + 1], line[j]
        n += 1

matrix = copy.deepcopy(generated_matrix)
        
t1 = time.time()     
for line in matrix:
    line = bubble_sort(line)

t2 = time.time()

bubble_sort_time = round((t2 - t1) * 1000)     
    
print_mat(matrix)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# сортировка вставками 

start_time = time.time()

def insertion_sort(line):
    for i in range(1, len(line)):
        for j in range(i, 0, -1):
            if line[j] < line[j - 1]:
                line[j], line[j - 1] = line [j - 1], line[j]
            else:
                break

matrix = copy.deepcopy(generated_matrix)

t1 = time.time()     

for line in matrix:
    line = insertion_sort(line)

t2 = time.time()

insertion_sort_time = round((t2 - t1) * 1000)   

print_mat(matrix)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
