```python
# –∫ –∑–∞–¥–∞–Ω–∏—é 1
def check_t1(t1):
    if not (1 <= t1 <= 15):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1 <= t1 <= 15')
    return 0
def check_n1(n1):
    if not (2 <= n1 <= 2000):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 2<=n1<=2000')
    return 0
def check_a1(a1):
    for item in a1:
        if not (0 <= item <= 230):
            raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 0<=item<=230')
    return 0
def calculations1(a1,n1,xor):
    for i in range(n1):
        xor ^= a1[i]
    if xor == 0:
        mess='YES'
    else:
        cnt = 0
        xr = 0
        for i in range(n1):
            xr ^= a1[i]
            if xor == xr:
                cnt += 1
                xr = 0
        if cnt >= 2:
            mess='YES'
        else:
            mess='NO'
    return mess
# –∫ –∑–∞–¥–∞–Ω–∏—é 2
def check_n2_k2(n2,k2):
    if not (1 <= n2 <= 100000):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1<=n2<=100000')
    if not (0 <= k2 <= n2):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 0<=k2<=n2')
    return 0
def calculations2(k2,c2,d2,b2,a2):
    for i in c2:
        d2[i] += 12
        if min(d2.values()) > k2:
            d2[c2[b2]] -= 1
            b2 += 1
        else:
            a2 += 1
        return a2

# –∫ –∑–∞–¥–∞–Ω–∏—é 3
def check_t3(t3):
    if not (1 <= t3 <= 10 ** 4):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1 <= t3 <= 10^4')
    return 0
def check_a3_b3(a3,b3):
    if not (1 <= a3 < b3 <= 10 ** 9):
        raise ValueError('1 <= a3<b3 <= 10^9')
    return 0
def calculations3(s3,b3,a3):

        while b3 > 0:
            s3 += b3 - a3
            a3 //= 10
            b3 //= 10
        return s3

# –∫ –∑–∞–¥–∞–Ω–∏—é 4
def check_t4(t4):
    if not (1 <= t4 <= 105):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1 <= t4 <= 105')
    return 0
def check_n4_k4(n4,k4):
    if not (1 <= k4 < n4 <= 2 * (10 ** 5)):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1 <= k4<n4 <= 2*(10^5)')
    return 0
def calculations4(a4,s4,n4,k4):
    for i in range(0, k4):
        b4 = [0] * 26
        for j4 in range(i, n4, k4):
            b4[ord(s4[j4]) - 97] += 1
            b4[ord(s4[k4 - 1 - j4]) - 97] += 1
        a4 += sum(b4) - max(b4)
    return a4

# –∫ –∑–∞–¥–∞–Ω–∏—é 8
def check_t8(t8):
    if not (1 <= t8 <= 10 ** 4):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1 <= t8 <= 10 ** 4')
    return 0
def check_n8(n8):
    if not (1 <= n8 <= 10 ** 9):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1<=n8<=10**9')
    return 0
def calculations8(n8):

        s8 = int((((n8 * 2 - 1) ** 0.5) - 1) / 2)
        return s8

# –∫ –∑–∞–¥–∞–Ω–∏—é 9
def check_n9(n9):
    if not (3 <= n9 <= 10 ** 5):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 3<=n9<=10**5')
    return 0
def check_ar9(ar9):
    if not (1 <= len(ar9) <= 10 ** 9):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1 <= len(ar9) <= 10**9')
    return 0
def calculations9(ar9,n9):
    ar9.sort()
    c9 = int(ar9[-1] ** (1 / (n9 - 1)) + 0.5)
    count9 = 0
    for i in range(n9):
        count9 += abs(ar9[i] - c9 ** i)
    return count9

# –∫ –∑–∞–¥–∞–Ω–∏—é 10
def check_p10(p10):
    if not (1 <= p10 <= 2 * (10 ** 4)):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1 <= p10 <= 2*(10**4)')
    return 0
def check_n10(n10):
    if not (1 <= n10 <= 2 * (10 ** 5)):
        raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 1<=n10<=2*(10**5)')
    return 0
def check_item(item):
    for item in item:
        if not (0 <= item <= 1):
            raise ValueError('–Ω–∞—Ä—É—à–µ–Ω—ã —É—Å–ª–æ–≤–∏—è 0<=l<=1')
    return 0
def calculations10(n10,c10,ans10,l10):
    for i in range(1, n10):
        if (l10[i] == 1):
            c10 += 1
        else:
            ans10 += c10 // 3
            c10 = 0
    ans10 += c10 // 3
    return ans10
```
